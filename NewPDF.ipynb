```python
from PIL import Image
import subprocess
import os
from google.colab import files

# Install Ghostscript
!apt-get install -y ghostscript

# Constants
MAX_FILE_SIZE_MB = 5 # Max file size allowed in MB

# Upload PDF file from local machine
uploaded = files.upload()

# Get the uploaded file name
uploaded_file_name = list(uploaded.keys())[0]
input_pdf = f"/content/{uploaded_file_name}"
compressed_pdf = f"/content/{uploaded_file_name.replace('.pdf', '_compressed.pdf')}"

# Check file size limit
def check_file_size(file_path, max_size_mb):
    """ Check if the file exceeds the max size limit """
    file_size_mb = os.path.getsize(file_path) / (1024 * 1024)  # Convert to MB
    if file_size_mb > max_size_mb:
        print(f"❌ ERROR: File size ({file_size_mb:.2f}MB) exceeds the {max_size_mb}MB limit.")
        return False
    print(f"✅ File size is within the allowed limit: {file_size_mb:.2f}MB")
    return True

# Compress a PDF file
def compress_pdf(output_file, file_to_compress):
    """ Compress a PDF file using Ghostscript """
    # Use quotes around file paths to handle spaces and special characters
    script = f"gs -dNOPAUSE -dBATCH -sDEVICE=pdfwrite -dCompatibilityLevel=1.4 -sOutputFile=\"{output_file}\" \"{file_to_compress}\""
    try:
        subprocess.run(script, shell=True, check=True)
        print(f"✅ Compression complete: {file_to_compress} → {output_file}")
        return output_file
    except FileNotFoundError as e:
        print(f"❌ Error compressing PDF: {e}")

# Check if the uploaded file meets the size limit
if check_file_size(input_pdf, MAX_FILE_SIZE_MB):
    compressed_file = compress_pdf(compressed_pdf, input_pdf)
    print(f"✅ Final Compressed PDF: {compressed_file}")

    # Trigger direct download of the compressed PDF file
    files.download(compressed_file)
else:
    print("❌ Process stopped due to file size exceeding the limit.")
```
